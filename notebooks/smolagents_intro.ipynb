```python
from smolagents import CodeAgent, HfApiModel, ToolCallingAgent, tool, DuckDuckGoSearchTool
```

```python
model = HfApiModel()
```

```python
@tool
def select_new_point(points: list[float]) -> float:
  """
  Given a list of sampled points, finds and returns new point.
  Args:
    points: A list with float format values of prior sampled points.
  Returns:
    New point.
  """
  ls = sorted(points)
  differences = [x-y for x,y in zip(ls[1:], ls[:-1])]
  idx = differences.index(max(differences))
  pt = (ls[idx] + ls[idx+1]) / 2
  return pt
```

```python
agent = CodeAgent(tools = [select_new_point],
                  model = model,
                  add_base_tools = True)
```

```python
agent.run("I am sampling points in the range 0 to 1. I have sampled points [0.1, 0.3, 0.4, 0.65, 0.9]. Suggest the next point to sample")
```
